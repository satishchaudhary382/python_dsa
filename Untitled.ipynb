```python
# Nepse data extraction 

class NEPSE():
  def __init__(self, ticker, start_date, end_date):
    self.ticker = ticker
    self.start_date = start_date
    self.end_date = end_date

  def scrape(self):
    url = "https://nepsealpha.com/nepse-data"
    head = header
    payload = {
    'symbol': self.ticker,
    'specific_date': '2022-08-06',
    'start_date': self.start_date,
    'end_date': self.end_date,
    'filter_type': 'date-range',
    '_token': token
}

    raw = httpx.post(url, data = payload, headers = head)
    df = pd.DataFrame(json.loads(raw.text)['data'])
    return df

  def data(self):
    raw_data = self.scrape()
    import warnings
    warnings.filterwarnings("ignore")
    raw_data = raw_data[['f_date','open','high','low','close','volume']]
    raw_data.columns = ['Date','Open','High','Low','Close','Volume']
    raw_data['Date'] = pd.to_datetime(raw_data['Date'])
    raw_data['Open'] = pd.to_numeric(raw_data['Open'])
    raw_data['High'] = pd.to_numeric(raw_data['High'])
    raw_data['Low'] = pd.to_numeric(raw_data['Low'])
    raw_data['Close'] = pd.to_numeric(raw_data['Close'])
    raw_data['Volume'] = pd.to_numeric(raw_data['Volume'])
    raw_data = raw_data.set_index("Date")
    return raw_data.sort_index(ascending =True)
```

```python
#@title Multiiprocessing function to download data

start_date = str(input("Enter start date (yyyy-mm-dd): "))
end_date = str(input("Enter end date (yyyy-mm-dd): "))
def extract(stock):
  # stock : Name of the stock or sector to be extracted
  # start_date : eg "2020-01-01"
  # end_date : eg "2022-01-01"

  return NEPSE(stock,start_date,end_date).data()


DATA = {}
def download(stock):
  results = None
  with multiprocessing.Pool(processes = 16 ) as pool:
    results = pool.map(extract,(stock))
    pool.close()
  for i,sec in zip(range(len(results)),stock):
    DATA[sec] = results[i]
  return DATA
```

```python
# Download stocks
import threading

DATA = {}

def extract(stock):
    # stock : Name of the stock or sector to be extracted
    # start_date : eg "2020-01-01"
    # end_date : eg "2022-01-01"
    return NEPSE(stock, start_date, end_date).data()

def download(stock):
    threads = []
    for s in stock:
        thread = threading.Thread(target=download_worker, args=(s,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    return DATA

def download_worker(stock):
    result = extract(stock)
    DATA[stock] = result
```

```python
#@title Class for sector Analysis

class sectorAnalysis:
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def absolute_strength(self, sector, index_string, length = 7, date = "2023-12-07"):
    f = {}
    as_df = self.dataframe.copy()
    temp_sector = set(sector)

    sectors_to_add = ["NEPSE", index_string]

    for sector_to_add in sectors_to_add:
        if sector_to_add not in temp_sector:
            temp_sector.add(sector_to_add)

    # Convert the set back to a list if needed
    temp_sector = list(temp_sector)

    for i in temp_sector:
      f[i] = as_df[i]['Close']#.pct_change(length).cumsum()
    new_f = {}
    for i in temp_sector:
      new_f[i] = f[i].values[-min([len(i) for i in f.values()]):]
    index_as_df = f['NEPSE'].keys()[-min([len(i) for i in f.values()]):]
    a = pd.DataFrame(new_f, index = index_as_df)
    a = a.loc[date:].pct_change().cumsum()
    # print(new_f)
    i,j = 0,0
    PLOTS_PER_ROW = 4
    fig, axs = plt.subplots(math.ceil(len(a.columns)/PLOTS_PER_ROW),PLOTS_PER_ROW, figsize = (20,10),sharex = True)
    fig.tight_layout()
    for col in a.columns:
      axs[i][j].plot(a[col]) # By default 50 is used to concentrate on last 50 days data
      axs[i][j].set_title(col)
      j+=1
      if j%PLOTS_PER_ROW ==0:
        i+=1
        j=0

  def relative_strength_pct_sector(self, sector, index_string, date = "2023-12-07"):
    f = {}
    rsp_df = self.dataframe.copy()

    temp_sector = set(sector)

    sectors_to_add = ["NEPSE", index_string]

    for sector_to_add in sectors_to_add:
        if sector_to_add not in temp_sector:
            temp_sector.add(sector_to_add)

    # Convert the set back to a list if needed
    temp_sector = list(temp_sector)
    # return temp_sector
    for i in temp_sector:
      f[i] = rsp_df[i]['Close']#.pct_change(length).cumsum()
    new_f = {}
    for i in temp_sector:
      new_f[i] = (f[i].values[-min([len(i) for i in f.values()]):] / f[index_string].values[-min([len(i) for i in f.values()]):])#.pct_change().cumsum()
    index_df = f['NEPSE'].keys()[-min([len(i) for i in f.values()]):]
    a = pd.DataFrame(new_f, index = index_df)
    a = a.loc[date:].pct_change().cumsum()

    # for i in temp_sector:
    #   f[i] = (self.dataframe[i]['Close']/self.dataframe[index_string]['Close']).pct_change().cumsum()
    # a = pd.DataFrame(f).dropna().fillna(method = 'bfill').iloc[:,1:]
    # # print(a)
    i,j = 0,0
    PLOTS_PER_ROW = 4
    fig, axs = plt.subplots(math.ceil(len(a.columns)/PLOTS_PER_ROW),PLOTS_PER_ROW, figsize = (20,10),sharex = True)
    fig.tight_layout()
    for col in a.columns:
      axs[i][j].plot(a[col])
      axs[i][j].set_title(col)
      j+=1
      if j%PLOTS_PER_ROW ==0:
        i+=1
        j=0

  def breadth_indicator(self, sector):
    sec = self.dataframe.copy()
    for i in sector:
      sec[i]['Symbol'] = i
      sec[i]['pct_change'] = sec[i]['Close'].pct_change()
      sec[i]['signs'] = np.sign(sec[i]['pct_change'])
    new_df = pd.concat(sec)
    new_df = new_df.reset_index()
    s = new_df.groupby(by = ['Date','signs'])['signs'].count().to_frame('count').reset_index()
    s = pd.pivot_table(s, values = 'count', index = 'Date', columns = 'signs').fillna(0)
    s = s.rename(columns = {1.0:"Advancers",0.0:"Neutral",-1.0:"Decliners"})
    s['percent'] = (s['Advancers']/(s['Advancers']+s['Decliners']))*100
    s['ema'] = ta.jma(ta.ema(s['percent'],20).dropna())
    s['ema_10'] = ta.jma(ta.ema(s['percent'],50).dropna())
    s['ema'].iloc[-200:].plot(figsize=(30,10))
    s['ema_10'].iloc[-200:].plot(figsize=(30,10))


nepse_sector = sectorAnalysis(sec)
# hydropower_sector = sectorAnalysis(hy)
finance_sector = sectorAnalysis(fin)
development_sector = sectorAnalysis(dev)
# manu_sector = sectorAnalysis(mn)
# micro_sector = sectorAnalysis(mic)
# other_sector = sectorAnalysis(oth)
# hotel_sector = sectorAnalysis(hot)
```
